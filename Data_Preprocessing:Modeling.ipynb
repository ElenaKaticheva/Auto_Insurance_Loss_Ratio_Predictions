```python
#Principal component analysis (PCA)
#Principal Component Analysis (PCA) is a linear dimensionality reduction technique that can be utilized for extracting information from a high-dimensional space by projecting it into a lower-dimensional sub-space. It tries to preserve the essential parts that have more variation of the data and remove the non-essential parts with fewer variation.
from sklearn.decomposition import PCA
pca = PCA().fit(X_train_scaled)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlim(0,7,1)
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance')
```

```python
#https://www.machinelearningplus.com/machine-learning/principal-components-analysis-pca-better-explained/
sklearn_pca=PCA()
X_Train=sklearn_pca.fit_transform(X_train_scaled)
sns.set(style='darkgrid')
f, ax = plt.subplots(figsize=(12, 12))
ax = sns.kdeplot(X_Train[:,0], X_Train[:,1], cmap="Greens",
          shade=True, shade_lowest=False)
ax = sns.kdeplot(X_Train[:,1], X_Train[:,2], cmap="Reds",
          shade=True, shade_lowest=False)
ax = sns.kdeplot(X_Train[:,2], X_Train[:,3], cmap="Blues",
          shade=True, shade_lowest=False)
red = sns.color_palette("Reds")[-2]
blue = sns.color_palette("Blues")[-2]
green = sns.color_palette("Greens")[-2]
ax.text(1, 0.5, "2nd and 3rd Projection", size=12, color=blue)
ax.text(-5, 4, "1st and 3rd Projection", size=12, color=red)
ax.text(4, -1.5, "1st and 2nd Projection", size=12, color=green)
plt.xlim(-12,12)
plt.ylim(-8,8)
plt.title('Principal component analysis', size=14)
```

```python
X_train = df.drop(['Loss_Amount'],axis=1)
y_train = df['Loss_Amount']
```

```python
# split the dataset into traning and testing data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, test_size=0.3, random_state=123)
print('dim(X_train) = ', X_train.shape)
print('dim(X_test) = ', X_test.shape)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import LinearRegression
from sklearn import metrics
# Linear Regression
model = LinearRegression()
model.fit(X_train, y_train)
LR_rmse = np.sqrt(mean_squared_error(y_test, model.predict(X_test)))
LR_mae = mean_absolute_error(y_test, model.predict(X_test))
print('Linear Regression - RMSE = ', LR_rmse)
print('Linear Regression - MAE = ', LR_mae)
```

```python
#Ridge Regression
from sklearn import linear_model
Ridge=linear_model.Ridge()
Ridge.fit(X_train,y_train)
DTR_rmse = np.sqrt(mean_squared_error(y_test, Ridge.predict(X_test)))
DTR_mae = mean_absolute_error(y_test, Ridge.predict(X_test))
print('Ridge Regression - RMSE = ', DTR_rmse)
print('Ridge Regression - MAE = ', DTR_mae)
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeRegressor
DTR = DecisionTreeRegressor()
DTR.fit(X_train, y_train)
DTR_rmse = np.sqrt(mean_squared_error(y_test, DTR.predict(X_test)))
DTR_mae = mean_absolute_error(y_test, DTR.predict(X_test))
print('Ridge Regression - RMSE = ', DTR_rmse)
print('Ridge Regression - MAE = ', DTR_mae)
```

```python
#XGBoost training
import xgboost as xgb

xg_trn_data = xgb.DMatrix(data = X_train, label =y_train) 
xg_test_data = xgb.DMatrix(data = X_test, label = y_test) 
  
# Parameter dictionary specifying base learner 
xgb_param = {"booster":"gblinear", "objective":"reg:linear"} 
  
bst = xgb.train(params = xgb_param, dtrain = xg_trn_data, num_boost_round = 10) 
xgb_rmse = np.sqrt(mean_squared_error(y_test, bst.predict(xg_test_data)))
xgb_mae = mean_absolute_error(y_test, bst.predict(xg_test_data))
print('XGBoost - RMSE = ', xgb_rmse)
print('XGBoost - MAE = ', xgb_mae)
```

```python
#RandomForestRegressor
from sklearn.ensemble import RandomForestRegressor
RFR = RandomForestRegressor(n_estimators = 50, max_depth = 4, n_jobs = -1, random_state=0)
RFR.fit(X_train, y_train)
RFR_rmse = np.sqrt(mean_squared_error(y_test, RFR.predict(X_test)))
RFR_mae = mean_absolute_error(y_test, RFR.predict(X_test))
print('Ridge Regression - RMSE = ', RFR_rmse)
print('Ridge Regression - MAE = ', RFR_mae)
```

```python
#KNN
from sklearn.neighbors import KNeighborsRegressor
knn_regressor = KNeighborsRegressor(5)
knn_regressor.fit(X_train, y_train)
knn_rmse = np.sqrt(mean_squared_error(y_test, knn_regressor.predict(X_test)))
knn_mae = mean_absolute_error(y_test, RFR.predict(X_test))
print('Ridge Regression - RMSE = ', knn_rmse)
print('Ridge Regression - MAE = ', knn_mae)
```

```python
testing_portfolios = 'testing_portfolios'
portfolios = [p for p in listdir(testing_portfolios) if isfile(join(testing_portfolios, p))]
```

```python
portfolio_list=[]
for index, portfolio in enumerate(portfolios):
    test = pd.read_csv(testing_portfolios+'/'+portfolio)
    test = test[columns[:len(columns)-1]]
    test = pd.get_dummies(test, columns = obj_columns, prefix_sep ='_')
    #print(test)
     
    #remove additional col
    for col in test.columns:
        if col not in df.columns and col !='Loss_Amount':
            test.drop(col, axis=1, inplace=True)
    #add the extra col
    for col in df.columns:
        if col not in test.columns and col !='Loss_Amount':
            test[col] = 0
    cols = []
    for col in df.columns.tolist():
        if col !='Loss_Amount':
            cols.append(col)
    test = test[cols]
    
    predictions = model.predict(test)
    values = np.log(sum(predictions)/(test['Annual_Premium']))
    portfolio_list.append(values)

portfolios = [folio.replace('test_', '') for folio in portfolios]
portfolios = [folio.replace('.csv', '') for folio in portfolios]
data = {'ID': portfolios, 'ln_LR': predictions}
df_new = pd.DataFrame.from_dict(data, orient='index')
print(df_new)

excel_exported = df_new.to_csv (r'ln_lr.csv',index = None, header=True)
    
   
```
